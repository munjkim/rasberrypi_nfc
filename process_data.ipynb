```python
import pandas as pd
import datetime
import numpy as np
```

```python
df_ID = pd.read_csv('./data/student.csv')
df_data = pd.read_csv('./data/reading.csv')
df = pd.DataFrame()
```

```python
df_ID
```

```python
df['Student ID'] = ['None'] * len(df_ID)
df['Name'] = ['None'] * len(df_ID)
df['UID'] = ['None'] * len(df_ID)

for i in range(len(df_ID)):
    sid = df_ID['Student ID'][i]
    name = df_ID.loc[df_ID['Student ID']==sid, 'Name'].values[0]

    try:
        uid = df_data.loc[df_data['Student ID']==sid, 'UID'].values[0]
    except IndexError:
        uid = None

    df['Student ID'][i] = sid
    df['Name'][i] = name
    df['UID'][i] = uid

    print(sid, name, uid)

df.to_csv('./data/data.csv')
```
