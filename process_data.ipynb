```python
import pandas as pd
import datetime
import numpy as np
```

```python
df_ID = pd.read_csv('./data/student.csv')
df_data = pd.read_csv('./data/reading.csv')
df = pd.DataFrame()
```

```python
# 초기화 (df 길이에 맞춰야 안전)
df = df_ID.copy()
df['UID'] = [None] * len(df)
df['last update'] = [None] * len(df)

for i in range(len(df_data)):
    student_id = df_data.iloc[i]['Student ID']

    df.loc[df['Student ID'] == student_id, 'UID'] = df_data.iloc[i]['UID']
    df.loc[df['Student ID'] == student_id, 'last update'] = df_data.iloc[i]['Date']

df.to_csv('./data/data.csv', index=False)
```
